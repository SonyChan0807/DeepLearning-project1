```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tr0 = np.load("normalized, Nd=1, h=25, dropout=0, fc=(25x2)/v1_tr0_loss.npy")
tr1 = np.load("normalized, Nd=1, h=25, dropout=0, fc=(25x2)/v1_tr1_loss.npy")
tr2 = np.load("normalized, Nd=1, h=25, dropout=0, fc=(25x2)/v1_tr2_loss.npy")
tr3 = np.load("normalized, Nd=1, h=25, dropout=0, fc=(25x2)/v1_tr3_loss.npy")
tr4 = np.load("normalized, Nd=1, h=25, dropout=0, fc=(25x2)/v1_tr4_loss.npy")

val0 = np.load("normalized, Nd=1, h=25, dropout=0, fc=(25x2)/v1_val0_loss.npy")
val1 = np.load("normalized, Nd=1, h=25, dropout=0, fc=(25x2)/v1_val1_loss.npy")
val2 = np.load("normalized, Nd=1, h=25, dropout=0, fc=(25x2)/v1_val2_loss.npy")
val3 = np.load("normalized, Nd=1, h=25, dropout=0, fc=(25x2)/v1_val3_loss.npy")
val4 = np.load("normalized, Nd=1, h=25, dropout=0, fc=(25x2)/v1_val4_loss.npy")
```

```python
plt.figure(figsize=(10,5))
plt.plot(tr0, label='k = 1')
plt.plot(tr1, label='k = 2')
plt.plot(tr2, label='k = 3')
plt.plot(tr3, label='k = 4')
plt.plot(tr4, label='k = 5')
plt.legend(loc='upper left')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Epoch", fontsize=20)
plt.ylabel("Training Loss", fontsize=20)
plt.savefig("LSTM Training Loss")
plt.show()

plt.figure(figsize=(10,5))
plt.plot(val0, label='k = 1')
plt.plot(val1, label='k = 2')
plt.plot(val2, label='k = 3')
plt.plot(val3, label='k = 4')
plt.plot(val4, label='k = 5')
plt.legend(loc='upper left')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Epoch", fontsize=20)
plt.ylabel("Validation Loss", fontsize=20)
plt.savefig("LSTM Validation Loss")
plt.show()
```
